#### Definindo as funções a serem usadas no Projeto

```python
def createBaseUser(id_user=5, nome='', telefone=00000000000, endereco='', dicio_usuarios=dict()):
    aux_dicio = {}
    
    aux_dicio['ID'] = id_user
    aux_dicio['Nome'] = nome
    aux_dicio['Telefone'] = telefone
    aux_dicio['Endereço'] = endereco
    aux_dicio['Status'] = True 

    dicio_usuarios[f'{id_user}'] = aux_dicio.copy()
    return dicio_usuarios


def createUser(dicio_usuarios=dict(), lista_ids=list()):
    dict_aux = {}
    id_user = len(lista_ids) + 1

    dict_aux['ID'] = id_user
    dict_aux['Nome'] = str(input('Qual o nome do usuário ?')).title().strip()
    dict_aux['Telefone'] = int(input('Qual o telefone do usuário? '))
    dict_aux['Endereço'] = str(input('Qual o endereço do usuário? ')).strip()
    dict_aux['Status'] = True 

    dicio_usuarios[f'{id_user}'] = dict_aux.copy()
    dict_aux.clear()
    lista_ids.append(id_user)
    
    

def checkUser(lista_ids=list(), dicio_usuarios=dict()):
    numero_de_id = str(input("Digite o número de ID que deseja validar? "))
    while int(numero_de_id) not in lista_ids or dicio_usuarios[numero_de_id]['Status'] == False:
        print("\033[31mUsuário não encontrado na base de dados!\033[m")
        numero_de_id = str(input("Qual o número de ID que deseja validar: "))
    else:
        print('Usuário validado com sucesso')
        return numero_de_id

    

def updateUser(id_user, dicio_usarios):    
    while True:
        print('''\nQual informação deseja atualizar?
        1 - Nome
        2 - Telefone
        3 - Endereço
        4 - Sair
        ''')
        menu = int(input("Qual opção deseja selecionar? "))
        while menu not in (1, 2, 3, 4):
            print("\033[31mOpção de atualização indisponível. Tente novamente.\033[m")
            menu = int(input("Qual opção deseja selecionar? "))

        if menu == 1:
            print('\n1')
            nome = str(input("Digite o nome: ")).strip().title()
            dicio_usarios[id_user]['Nome'] = nome
        elif menu == 2:
            print('\n2')
            telefone = int(input("Digite o número de telefone: "))
            dicio_usarios[id_user]['Telefone'] = telefone
        elif menu == 3:
            print('\n3')
            endereco = str(input('Digite o endereço: ')).strip().title()
            dicio_usarios[id_user]['Endereço'] = endereco
        else:
            break
    


def getUser(dicio_usuarios = dict(), lista_ids=list()):
    id_do_user = str(input('Digite o número de ID que deseja exibir? '))
    while not (int(id_do_user) in lista_ids and dicio_usuarios[id_do_user]['Status'] == True):
        print("\033[31mOpção inválida. Tente novamente.\033[m")
        id_do_user = str(input('Digite o número de ID que deseja exibir? '))
    else:
        for chave, valor in dicio_usuarios[id_do_user].items():
            print(f'A chave {chave} é {valor}') 



def getUsers(dicio_usuarios=dict(), lista_ids=list()):
    while True:
        for x in lista_ids:
            if dicio_usuarios[str(x)]['Status'] == True:
                for chave, valor in dicio_usuarios[str(x)].items():
                    print(f'{chave}: {valor}')
        break
```

#### Adicionando pontos chaves do Projeto

```python
total_users = dict()
lista_id = [1, 2, 3, 4, 5]
createBaseUser(1, 'Mokey D. Luffy', 111111111, 'East Blue', total_users)
createBaseUser(2, 'Roronoa Zoro', 22222222222, 'East Blue', total_users)
createBaseUser(3, 'God Ussop', 33333333333, 'East Blue', total_users)
createBaseUser(4, 'Vinsmoke Sanji', 44444444444, 'East Blue', total_users)
createBaseUser(5, 'Nami', 5555555555, 'East Blue', total_users)
```

#### Corpo do Projeto em si

```python
while True:
    print()
    print(f"{'Boas Vindas ao Nosso Sistema de Cadastro'}\n")
    print("""Você possui 6 opções para escolher:\n
    1 - Inserir Usuário
    2 - Excluir Usuário
    3 - Atualizar Usuário
    4 - Informações de um Usuário
    5 - Informações de todos os Usuários
    6 - Sair\n
    """)

    opcao = int(input("Escolha uma opção acima em número para prosseguir: "))

    while opcao not in range(1,7):
        print("\033[31mOpção inválida. Tente novamente.\033[m")
        opcao = int(input("Escolha uma opção acima para prosseguir: "))

    if opcao == 1:
        createUser(total_users, lista_id)
        print()

    elif opcao == 2:
        num_id = checkUser(lista_id, total_users)
        total_users[num_id]['Status'] = False
        print(f'\033[35mO usuário de ID {num_id} foi excluído com sucesso\033[m')
        print()
        
    elif opcao == 3:
        num_id = checkUser(lista_id, total_users)
        updateUser(num_id, total_users)
        print('\033[34mUsuário atualizado com sucesso.\033[m')
        print()
        
    elif opcao == 4:
        getUser(total_users, lista_id)
        print()

    elif opcao == 5:
        getUsers(total_users, lista_id)
        print()
    else:
        print('Obrigado por usar nosso sistema de cadastro')
        break
```
